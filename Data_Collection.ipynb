# Data Collection

1. Dữ liệu liên quan đến chủ đề gì?

- Chủ đề (Topic), Lĩnh vực (Domain), Hiện tượng (Phenomenon):

    - Chủ đề (Topic): Đánh giá hạnh phúc và phúc lợi của con người trên quy mô toàn cầu (Global Happiness and Well-being).

    - Lĩnh vực (Domain): Khoa học Xã hội (Social Science), Kinh tế học (cụ thể là Kinh tế học Hạnh phúc - Happiness Economics) và Thống kê ứng dụng.

    - Hiện tượng (Phenomenon): Dữ liệu mô tả hiện tượng chênh lệch về mức độ hài lòng cuộc sống giữa các quốc gia khác nhau và cách các yếu tố như kinh tế, sức khỏe, tự do cá nhân tác động đến cảm nhận hạnh phúc của con người.

- Ngữ cảnh thực tế (real-world context):

    - Bộ dữ liệu này đại diện cho Báo cáo Hạnh phúc Thế giới (World Happiness Report). Đây là một cuộc khảo sát mang tính bước ngoặt về trạng thái hạnh phúc toàn cầu. Nó xếp hạng 156 quốc gia dựa trên mức độ hạnh phúc của người dân, được đánh giá qua các chỉ số như GDP bình quân đầu người, hỗ trợ xã hội, tuổi thọ kỳ vọng, quyền tự do, sự hào phóng và mức độ tham nhũng. Dữ liệu này phản ánh sự quan tâm ngày càng tăng của các chính phủ và tổ chức trong việc sử dụng chỉ số hạnh phúc để hoạch định chính sách công.

2. Nguồn của dữ liệu là gì?

- Nền tảng: `Kaggle.com`.

- URL: [World Happiness Report](https://www.kaggle.com/datasets/unsdsn/world-happiness)

- Tác giả và tổ chức:

    - Tổ chức: Sustainable Development Solutions Network (Mạng lưới Giải pháp Phát triển Bền vững của Liên Hợp Quốc - UN SDSN).

    - Dữ liệu gốc được lấy từ Gallup World Poll.

- Năm xuất bản / thu thập:

    - Trên Kaggle, bộ dữ liệu này bao gồm các file cho các năm báo cáo: 2015, 2016, 2017, 2018, 2019. Nhóm quyết định chọn bản mới nhất `2019`.

3. Dữ liệu có giấy phép để sử dụng?

- Giấy phép (License): `CC0: Public Domain` (Phạm vi công cộng).

- Được phép sử dụng cho mục đích giáo dục. Giấy phép CC0 cho phép sao chép, sửa đổi, phân phối và thực hiện công việc, ngay cả cho mục đích thương mại mà không cần xin phép.

- Các hạn chế sử dụng hoặc yêu cầu ghi nguồn:

    - Hạn chế: Không có hạn chế pháp lý đáng kể nào dưới giấy phép `CC0`.

    - Ghi nguồn: Mặc dù `CC0` không bắt buộc về mặt pháp lý, nhưng về mặt học thuật và đạo đức nghiên cứu, nên trích dẫn nguồn từ "Sustainable Development Solutions Network" và "Gallup World Poll" khi sử dụng.

4. Dữ liệu được thu thập như thế nào?

- Phương pháp thu thập (Collection method):

    - Khảo sát (Survey): Dữ liệu điểm số hạnh phúc chủ yếu dựa trên Gallup World Poll.

    - Câu hỏi chính được sử dụng là `thang đo Cantril` (Cantril ladder): Người tham gia được yêu cầu tưởng tượng về một chiếc thang, với bậc 10 là cuộc sống tốt đẹp nhất và bậc 0 là tồi tệ nhất, sau đó đánh giá cuộc sống hiện tại của họ đang ở bậc nào.

- Đối tượng và cách lấy mẫu:

    - Đối tượng: Người dân tại hơn 155 quốc gia trên thế giới.

    - Lấy mẫu: Sử dụng mẫu đại diện quốc gia (nationally representative samples). Thông thường, điểm số được lấy từ các mẫu đại diện toàn quốc trong những năm (VD: 2013-2016) và sử dụng trọng số Gallup để đưa ra các ước tính mang tính đại diện

- Khoảng thời gian thu thập:

    - Dữ liệu trong các file csv qua các năm (ví dụ 2015, 2016...) là kết quả báo cáo của năm đó. Điểm số hạnh phúc thường là giá trị trung bình của các cuộc khảo sát trong 3 năm gần nhất để đảm bảo độ ổn định và tăng độ chính xác thống kê.

- Các hạn chế hoặc thiên kiến:

    - Tính chủ quan: Dữ liệu dựa trên đánh giá tự cảm nhận (self-reported) của người dân, có thể bị ảnh hưởng bởi văn hóa hoặc tâm trạng tại thời điểm khảo sát.

    - Dữ liệu khuyết thiếu: Một số chỉ số phụ (như GDP, hỗ trợ xã hội) có thể được lấy từ các nguồn khác (như World Bank, WHO) và có thể có sai số hoặc ước tính bổ sung cho các quốc gia thiếu số liệu chính thức.

    - Biến `Dystopia`: Đây là một quốc gia giả tưởng có chỉ số thấp nhất thế giới dùng làm mốc so sánh, đây là một khái niệm thống kê chứ không phải dữ liệu thực tế.


5. Tại sao lại chọn dữ liệu này?

- Điều mà nhóm quan tâm trong chủ đề này:

    - Sự tò mò về việc yếu tố nào thực sự định hình nên hạnh phúc của một quốc gia: Là tiền bạc (GDP), sức khỏe (tuổi thọ), hay sự tự do và lòng tốt?

    - Sự chênh lệch chất lượng cuộc sống giữa các khu vực khác nhau trên thế giới (Châu Âu, Châu Á, Châu Phi...)

    - Mong muốn tìm hiểu xem Việt Nam đứng ở đâu trên bản đồ hạnh phúc thế giới.

- Câu hỏi hoặc thông tin chi tiết tiềm năng mà dữ liệu có thể cung cấp / trả lời:

    - GDP cao thì có đi với việc hạnh phúc cao không?

    - Tại sao một số nước Bắc Âu (Phần Lan, Đan Mạch...) luôn đứng đầu bảng xếp hạng?

    - Mối tương quan giữa Tham nhũng (Corruption) và Hạnh phúc là gì?
